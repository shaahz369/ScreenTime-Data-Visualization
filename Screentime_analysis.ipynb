```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel("Screentime-App-Details.xlsx")
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df["Date"] = pd.to_datetime(df["Date"])
plt.bar(df["Date"],df["Usage"])
plt.xlabel("date")
plt.ylabel("usage")
plt.show()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df["Date"] = pd.to_datetime(df["Date"])
plt.bar(df["Date"],df["Notifications"])
plt.xlabel("date")
plt.ylabel("Notifications")
plt.show()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df["Date"] = pd.to_datetime(df["Date"])
plt.bar(df["Notifications"],df["Usage"],color="green")
plt.xlabel("Notifications")
plt.ylabel("usage")
plt.show()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df["Date"] = pd.to_datetime(df["Date"])
plt.scatter(df["Notifications"],df["Usage"],color="green")
plt.xlabel("Notifications")
plt.ylabel("usage")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("date")
plt.ylabel("usage")
plt.title("USAGE")
for app, data in df.groupby("App"):
    plt.bar(data["Date"], data["Usage"], label=app)

plt.legend(title="App")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df["Date"] = pd.to_datetime(df["Date"])
plt.bar(df["Date"],df["Times opened"])
plt.xlabel("date")
plt.ylabel("Times opened")
plt.show()
```

```python
plt.xlabel("Date")
plt.ylabel("Notifications")
plt.title("NOTIFICATION")
for app, data in df.groupby("App"):
    plt.bar(data["Date"], data["Notifications"], label=app)

plt.legend(title="App")
plt.show()
```

```python
plt.xlabel("Date")
plt.ylabel("Times opened")
plt.title("TIMES OPEND")
for app, data in df.groupby("App"):
    plt.bar(data["Date"], data["Times opened"], label=app)

plt.legend(title="App")
plt.show()

```
